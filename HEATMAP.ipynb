```python
import geopandas as gpd

lotes_comercio = gpd.read_file('./data/LOTES/lotes_completo.shp')  
# convert it to the projection of our folium openstreetmap  
lotes_comercio = lotes_comercio
```

```python
lotes_comercio['centroid'] = lotes_comercio.geometry.centroid
```

```python
lotes_comercio['Longitude'] = lotes_comercio.geometry.centroid.x.astype(float)
lotes_comercio['Latitude'] = lotes_comercio.geometry.centroid.y.astype(float)

points = lotes_comercio[['Latitude', 'Longitude']]
```

```python
import folium
from folium.plugins import HeatMap

mapa = folium.Map([-07.15, -34.85],
                  zoom_start=12,
                  tiles='cartodbpositron')

heat_data = [[row['Latitude'],row['Longitude']] for index, row in points.iterrows()]
HeatMap(heat_data,  max_val=1, radius=2, blur=2, overlay=True).add_to(mapa)
mapa
```
