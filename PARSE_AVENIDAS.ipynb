```python
import geopandas as gpd
import os
import shapely
import subprocess
from lib.parse import *
from lib.funcs import *
import fiona
```

```python
# Enable fiona driver
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

# Read file
avenidas_df = gpd.read_file('./data/BASES/AVENIDAS/avenidas.kml', driver='KML')

# Drop Z dimension of polygons that occurs often in kml 
avenidas_df.geometry = avenidas_df.geometry.map(lambda polygon: shapely.ops.transform(lambda x, y, z: (x, y), polygon))
```

```python
avenidas_df = avenidas_df.join(avenidas_df.apply(parse_html, axis=1))
avenidas_df = avenidas_df.drop('Description', 1)
```

```python
avenidas_df = avenidas_df.to_crs(epsg=3857)
```

```python
avenidas_df.to_file("./data/BASES/AVENIDAS/avenidas.shp")
```
